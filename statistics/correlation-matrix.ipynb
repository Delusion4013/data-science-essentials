## Corrleation-matrix

A correlation matrix isÂ a table that displays the correlation coefficient between multiple variables.

The values ranges from -1 to +1, indicating the strength and direction of the linear association

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = [215, 325, 185, 332, 406, 522, 412,
     614, 544, 421, 445, 408],
y = [14.2, 16.4, 11.9, 15.2, 18.5, 22.1, 
     19.4, 25.1, 23.4, 18.1, 22.6, 17.2]
matrix = np.corrcoef(x, y)
print(matrix)
```

```python
# Create a heatmap of the correlation matrix
plt.figure(figsize=(4, 3))
sns.heatmap(matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
import pandas as pd

# Sample DataFrame
data = {
    'x': [45, 37, 42, 35, 39],
    'y': [38, 31, 26, 28, 33],
    'z': [10, 15, 17, 21, 12]
}

df = pd.DataFrame(data)

# Calculate the correlation matrix
correlation_matrix = df.corr()
print(correlation_matrix)
```

```python
# Create a heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()
```
